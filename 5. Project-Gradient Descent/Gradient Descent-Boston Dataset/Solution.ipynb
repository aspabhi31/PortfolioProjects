```python
train_data=np.loadtxt('train.csv', delimiter=',')
```

```python
def run():
  learning_rate=0.0001
  num_iterations=1000
  m, c=gd(np.array(train_data), learning_rate, num_iterations)
  return m,c
```

```python
def step_gradient(points, learning_rate, m, c):
  m_slope=[0 for i in range(len(points[0, 0:-1]))]
  c_slope=0
  M=len(points)
  N=len(points[0, 0:-1]) 
  for i in range(M):
    for j in range(N):
      x=points[i][j]
      y=points[i][-1]
      m_slope[j]+=-2*x*(y-m[j]*x-c)/M
      c_slope+=-2*(y-m[j]*x-c)/M
  m_slope=np.array(m_slope)
  m=np.array(m)
  new_m=m-learning_rate*m_slope
  new_c=c-learning_rate*c_slope
  return new_m, new_c
```

```python
def cost(points, m, c):
  total_cost=0
  M=len(points)
  N=len(points[0, 0:-1])
  for i in range(M):
    for j in range(N):
      x=points[i][j]
      y=points[i][-1]
      total_cost+=(1/M)*((y-m[j]*x-c)**2)
  return total_cost
```

```python
def gd(points, learning_rate, num_iterations):
  m=[0 for i in range(len(points[0, 0:-1]))]
  c=0
  for i in range(num_iterations):
    m,c=step_gradient(points, learning_rate, m, c)
    print("cost: ", cost(points, m, c))
  return m, c
```
